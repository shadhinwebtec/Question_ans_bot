```python
import torch
import pandas as pd 
from transformers import pipeline
```

```python
demo_model=pipeline(task='table-question-answering',model="google/tapas-base-finetuned-wtq",device=0)
```

```python
table=pd.read_csv(r"/kaggle/input/cricket/Cricket.csv",encoding='latin1')
```

```python
table = table.astype(str)
```

```python
table
```

```python
query = "Who has scored the highest runs?"
print(demo_model(table=table,query=query)["answer"])
```

```python
demo_model.model.save_pretrained("tapas_saved_model")
demo_model.tokenizer.save_pretrained("tapas_saved_model")
```
